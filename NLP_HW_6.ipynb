```python
Best_books_Amazon = ["The Midnight Library: A Novel",
                     "Where the Crawdads Sing",
                     "Billy Summers",
                     "I Am Watching You",
                     "The Wife's House: An absolutely unputdownable psychological thriller",
                     "The Giver of Stars: A Novel",
                     "Devoted",
                     "The Guardians: A Novel",
                     "Americanon: An Unexpected U.S. History in Thirteen Bestselling Books",
                     "The Best is Yet to Come: The new delightfully uplifting and life-affirming novel about love, friendship and second chances in 2021",
                     "The Write Hook: A Paranormal Women’s Fiction Novel: My So-Called Mystical Midlife Book One",
                     "Home (Oprah's Book Club): A Novel",
                     "The Book Woman of Troublesome Creek: A Novel",
                     "Long Road to Mercy (An Atlee Pine Thriller Book 1)",
                     "The Other Mrs.: A Novel",
                     "The Sleep Experiment: An edge-of-your-seat psychological thriller (World's Scariest Legends)",
                     "Cold Silence: A Chilling Psychological Thriller",
                     "The Saturday Morning Park Run: The most gloriously uplifting and page-turning fiction book of the 2021!",
                     "All The Light We Cannot See (Thorndike Reviewers' Choice)",
                     "Freckled: A Memoir of Growing up Wild in Hawaii (Memoir Series Book 1)",
                     "Little Fires Everywhere: A Novel",
                     "The Shadow Box",
                     "The Girl Who Lived: A Thrilling Suspense Novel",
                     "Wool"]
print("Our list is made up of",len(Best_books_Amazon),"books")
```

## Part B

```python
%%time
import itertools
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

Pair_list = []
Cosine_value = []

for pair in itertools.product(Best_books_Amazon, repeat=2):
    Pair_list.append(pair)
    X_list = word_tokenize(pair[0]) 
    Y_list = word_tokenize(pair[1])
    
    sw = stopwords.words('english') 
    l1 =[];l2 =[]
    X_set = {w for w in X_list if not w in sw} 
    Y_set = {w for w in Y_list if not w in sw}
    rvector = X_set.union(Y_set) 
    for w in rvector:
        if w in X_set: l1.append(1) # create a vector
        else: l1.append(0)
        if w in Y_set: l2.append(1)
        else: l2.append(0)
    c = 0
    for i in range(len(rvector)):
        c+= l1[i]*l2[i]
    cosine = c / float((sum(l1)*sum(l2))**0.5)
    Cosine_value.append(cosine)
    
Amazon_results = pd.DataFrame({'Pair':Pair_list,'Cosine_Score':Cosine_value},columns=['Pair','Cosine_Score'])

Amazon_results
```

## Part C

```python
rslt_df = Amazon_results.loc[Amazon_results['Cosine_Score'] > .6] 
    
rslt_df.sort_values(by=['Cosine_Score'])
    
```

```python
%%time
New_Pair = []
Cosine_score = []

for x in Best_books_Amazon:
    for y in google_results_list:
        New_Pair.append((x+"|"+y))
        X_list = word_tokenize(x) 
        Y_list = word_tokenize(y)
    
        sw = stopwords.words('english') 
        l1 =[];l2 =[]
        X_set = {w for w in X_list if not w in sw} 
        Y_set = {w for w in Y_list if not w in sw}
        rvector = X_set.union(Y_set) 
        
        for w in rvector:
            if w in X_set: l1.append(1) # create a vector
            else: l1.append(0)
            if w in Y_set: l2.append(1)
            else: l2.append(0)
        c = 0
        
        for i in range(len(rvector)):
            c+= l1[i]*l2[i]
        cosine = c / float((sum(l1)*sum(l2))**0.5)
        Cosine_score.append(cosine)
    
google_results = pd.DataFrame({'Pair':New_Pair,'Cosine_Score':Cosine_score},columns=['Pair','Cosine_Score'])

google_results
        
```
